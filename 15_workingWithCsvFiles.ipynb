opening a csv file from an URL

```python
import pandas as pd
import requests
from io import StringIO

url = "http://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Fireforx/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

Sep parameter : The sep parameter in pd.read_csv() specifies the delimiter that separates the columns in the file. By default, it's set to a comma (,), but if you're working with a tab-separated values file (like a .tsv), you use sep='\t', where \t represents a tab character.              

A delimiter is a character or sequence of characters used to separate values in a text file :    
CSV(,)    
TSV(\t)

```python
pd.read_csv('./DataSets/15_weather.tsv', sep='\t').head()

# If data colonm name is not defined then name
# pd.read_csv('./DataSets/15_weather.tsv', sep='\t' names=['MaxTemp', 'Rainfall', etc all column names])
```

index_col parameter : convert the column into index

```python
pd.read_csv('weather.tsv', sep='\t', index_col='MinTemp')
```

Header parameter : if your header treat as a rows

```python
pd.read_csv('weather.tsv', sep='\t', header=1)
```

use_cols parameter :  if you only specific columns for filtering 

```python
pd.read_csv('weather.tsv', sep='\t', usecols=['MinTemp', 'MaxTemp', 'Rainfall'])
```

skiprows/nrows parameter 

```python
pd.read_csv('Weather.tsv', sep='\t', skiprows=[1,5])
```

nrows : 100 rows only show

```python
pd.read_csv('Weather.tsv', sep='\t', nrows=100)
```

converters : i want to short the name (FIrst Name to (Name))

```python
def rename(name):
    if name == "Callahan":
        return "ch"
    else:
        return "Not change"
```

```python
rename('Callahan')
```

```python
pd.read_csv('customersSaledata-10000.csv', converters={'Last Name':rename})
```
