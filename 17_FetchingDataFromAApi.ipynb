API : Application programming interface           
It's a set of rules that allows different software applications to communicate with each other.   
APIs typically work using   protocols (like GET, POST, PUT, DELETE) and often return data in a standard format like JSON or XML.    
Examples of APIs :   

REST APIs: Representational State Transfer APIs are widely used for web services. They typically use HTTP requests to interact with resources.     
HTTP : stands for Hypertext Transfer Protocol, and it's the protocol used for transmitting data over the web. It's the foundation of any data exchange on the Web and is used by web browsers and servers to request and send web pages, images, videos, and other content.           

Protocol: A set of rules or standards that define how data is transmitted between a client (like a web browser) and a server (where a website is hosted).   
Text-Based Communication: HTTP allows clients (e.g., browsers) to request information from a server, and the server responds with the requested data, such as HTML pages or media.        
HTTP Request: Sent by the client (e.g., your web browser).      
Request Line: Contains the HTTP method (GET, POST, etc.), the requested URL, and the HTTP version (e.g., HTTP/1.1). Example: GET /index.html HTTP/1.1

I want to do it in a loop, and the loop will run as many times as there are pages.     
Each time, a new dataframe will be created inside the loop, and it will be appended to the full data.    
Then, create the 8k movie data.
We will ignore the index, that's why the loop will run again, and the same index will be repeated. By using the index, it will ignore the previous index.  

```python
# Empty DataFrame
df = pd.DataFrame()
```

```python
for i in range(1,498):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
    df = pd.concat([df, temp_df], ignore_index=True)
```
