Feature engineering is a crucial step in the machine learning pipeline. It involves transforming raw data into features that make machine learning algorithms more effective.  
Here are the key steps in feature engineering:   

Feature engineering can indeed take up a significant portion of a machine learning projectâ€”often around 30% or even more of the total project time.           
Feature engineering is not just a process; it's more like an art or a problem-solving technique. Everyone has their own unique approach to performing feature engineering.           
 
"features" typically refer to the input variables or columns in your dataset.    
It is said about feature engineering that if you provide a poor algorithm with excellent features and a powerful algorithm with poor features, the one with the poor features will perform worse.

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('23_Social_Network_Ads.csv')
df = df.iloc[:, 2:]
df.sample(5)
```

If you want to Feature scaling, Then you should always train and split, it is recommended.       
df.drop('Purchased', axis=1) :: That means it will leave 'Purchased' column and put the 'age' and 'salary' in the 'x_train'.     
random_state=0 ensures that the same 30% of data will be used for the test set every time this line is executed.            

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Purchased', axis=1), df['Purchased'], test_size=0.3, random_state=0)


X_train.shape, X_test.shape
```

Standard Scalar class :  will take each value and transform it, using formula.     
If you want, you can create the class yourself, but it's given. then why gan. masti.    

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# fit the scalar to the train set, it will learn the parameter, fit means calculate both column (age, salary) mean and SD of X_train.  
scaler.fit(X_train)

# transforms train and test sets 
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
scaler.mean_
```

convert X_train_scale into dataframe    

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)

X_train_scaled.head()
```

Effect of Scaling

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,5))

ax1.scatter(X_train['Age'], X_train['EstimatedSalary'])
ax1.set_title("Before Scaling")

ax2.scatter(X_train_scaled['Age'], X_train_scaled['EstimatedSalary'], color='red')
ax2.set_title("After Scaling")

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,5))

ax1.set_title("Before scaling")
sns.kdeplot(X_train['Age'], ax=ax1)
sns.kdeplot(X_test['EstimatedSalary'], ax=ax1)

ax2.set_title("After scaling")
sns.kdeplot(X_train_scaled['Age'], ax=ax2)
sns.kdeplot(X_test_scaled['EstimatedSalary'], ax=ax2)
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=2, figsize=(12,5))

ax1.set_title('Salary Distribution Before Scaling')
sns.kdeplot(X_train['EstimatedSalary'], ax=ax1)

ax2.set_title('Salary Distribution After standard Scaling')
sns.kdeplot(X_train_scaled['EstimatedSalary'], ax=ax2)

plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr_scaled = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
lr_scaled.fit(X_train_scaled, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred_scaled = lr_scaled.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
print("Actual", accuracy_score(y_test, y_pred))
print('scaled', accuracy_score(y_test, y_pred_scaled))
```
