1. Simple LR        
Simple Linear Regression (SLR) is a method used to model the relationship between two continuous variables. One variable is dependent (target variable), and the other is independent (predictor variable). The relationship between these variables is assumed to be linear, meaning that a straight line can best describe how the independent variable affects the dependent variable.       
Basic Idea :      
In simple linear regression, we fit a straight line through the data points that minimizes the sum of squared errors (the vertical distance between the data points and the line). This line can be used for predictions.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
```

```python
plt.scatter(df['cgpa'], df['package'])
plt.xlabel('CGPA')
plt.ylabel("Package(in LPA)")
```

```python
x = df.iloc[:, 0:1]
y = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
X_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
# fit work is to train the model
lr.fit(X_train, y_train)
```

```python
x_test
```

```python
y_test
```

predict : I will give this information to my model and the model will predict this.

```python
print(lr.predict(x_test.iloc[0].values.reshape(1,1)))
print((lr.predict(x_test.iloc[1].values.reshape(1,1))))
print((lr.predict(x_test.iloc[2].values.reshape(1,1))))
```

```python
# a line is drow really, this is best fit line.
plt.scatter(df['cgpa'], df['package'])
plt.plot(X_train, lr.predict(X_train), color='red')
plt.xlabel('CGPA')
plt.ylabel("Package(in LPA)")
```

```python
# Slop value 
m = lr.coef_

# Intercept 
b = lr.intercept_

m,b
```
